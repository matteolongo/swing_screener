Momentum measures medium-term performance.

We compute:
- `mom_6m`: 6 month return (approx 126 trading days)
- `mom_12m`: 12 month return (approx 252 trading days)

Relative Strength (RS) compares vs a benchmark (default: **SPY**):
- `rs_6m = mom_6m(stock) - mom_6m(SPY)`

Positive RS means the asset outperformed the market over the lookback.

```python
from swing_screener.data.market_data import fetch_ohlcv, MarketDataConfig
from swing_screener.indicators.momentum import compute_momentum_features, MomentumConfig

tickers = ["AAPL","MSFT","NVDA","SPY"]
ohlcv = fetch_ohlcv(tickers, MarketDataConfig(start="2022-01-01"))

mom = compute_momentum_features(ohlcv, MomentumConfig(benchmark="SPY"))
mom.sort_values("rs_6m", ascending=False).head(10)
```
