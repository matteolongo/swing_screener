```python
#!/usr/bin/env python3
"""Universe Basics - Loading and filtering ticker universes."""

from swing_screener.data.universe import (
    list_package_universes,
    load_universe_from_package,
    load_universe_from_file,
    filter_ticker_list,
    apply_universe_config,
    UniverseConfig,
    save_universe_file,
)
```

```python
# List all available packaged universes
universes = list_package_universes()
print(f"Available universes ({len(universes)}):")
for u in universes[:10]:
    print(f"  - {u}")
if len(universes) > 10:
    print(f"  ... and {len(universes) - 10} more")

print("\n" + "=" * 50 + "\n")
```

```python
# Load a universe from the package
tickers = load_universe_from_package("usd_mega_stocks")
print(f"Loaded usd_mega_stocks: {len(tickers)} tickers")
print(f"First 5: {tickers[:5]}")

print("\n" + "=" * 50 + "\n")
```

```python
# Load with configuration (cap at 50 tickers, ensure SPY is included)
cfg = UniverseConfig(benchmark="SPY", ensure_benchmark=True, max_tickers=50)
tickers_limited = load_universe_from_package("usd_all", cfg)
print(f"Loaded usd_all (capped at 50): {len(tickers_limited)} tickers")
print(f"Benchmark included: {'SPY' in tickers_limited}")
print("\n" + "=" * 50 + "\n")
```

```python
# Filter a ticker list
all_tickers = load_universe_from_package("usd_all")
filtered = filter_ticker_list(
    all_tickers,
    include=["AAPL", "MSFT"],  # Always include these
    exclude=["SPY", "QQQ"],  # Exclude these
    grep="INT",  # Only tickers containing "DEF"
)
print(f"Filtered tickers: {filtered}")

print("\n" + "=" * 50 + "\n")
```

```python
# Save a custom universe to file
custom_tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META"]
save_universe_file(custom_tickers, "my_universe.csv")
print("Saved custom universe to my_universe.csv")
```
