```python
from swing_screener.data.market_data import MarketDataConfig, fetch_ohlcv
```

## Step 1 - Configure default fetch behavior

This matches the original script using local cache and adjusted prices.


```python
config = MarketDataConfig(
    start="2024-01-01",
    end="2024-12-31",
    auto_adjust=True,
    progress=False,
    cache_dir=".cache/market_data",
)

tickers = ["AAPL", "MSFT", "GOOGL"]
print(f"Fetching OHLCV for {tickers}...")
```

## Step 2 - Fetch and inspect the OHLCV frame

Validate shape, coverage, and available price fields.


```python
            df = fetch_ohlcv(tickers, cfg=config)

            print(f"DataFrame shape: {df.shape}")
            print(f"Date range: {df.index.min()} to {df.index.max()}")
            print(f"Columns: {df.columns.get_level_values(0).unique().tolist()}")

            close_prices = df["Close"]
            print("
Latest close prices:")
            close_prices.tail()
```

## Step 3 - Try alternate caching modes

Use automatic end-date, force refresh, and cache bypass paths.


```python
config_live = MarketDataConfig(start="2024-01-01")
df_live = fetch_ohlcv(["NVDA"], cfg=config_live)
print(f"Live data (auto end): {df_live.index[-1]}")

df_fresh = fetch_ohlcv(["AAPL"], cfg=config, force_refresh=True)
print(f"Force refreshed data for AAPL: {len(df_fresh)} rows")

df_no_cache = fetch_ohlcv(
    ["TSLA"],
    cfg=MarketDataConfig(start="2024-06-01"),
    use_cache=False,
)
print(f"Fetched without cache: {len(df_no_cache)} rows")
```
