```python
from swing_screener.data.providers import (
    YfinanceProvider,
    get_default_provider,
    get_market_data_provider,
)
```

## Step 1 - Use default provider selection

By default this resolves to yfinance in a standard setup.


```python
provider = get_default_provider()
print(f"Provider: {provider.get_provider_name()}")

df = provider.fetch_ohlcv(
    ["AAPL", "MSFT", "NVDA"],
    start_date="2024-01-01",
    end_date="2024-12-31",
)
print(f"Fetched {len(df)} rows")
df["Close"].tail(3)
```

## Step 2 - Use `YfinanceProvider` directly

Direct use exposes helper methods for latest price, ticker info, and market status.


```python
yf_provider = YfinanceProvider(
    cache_dir=".cache/market_data",
    auto_adjust=True,
    progress=False,
)

latest_price = yf_provider.fetch_latest_price("AAPL")
info = yf_provider.get_ticker_info("MSFT")

print(f"AAPL latest price: ${latest_price}")
print(f"MSFT info: {info}")
print(f"Market open: {yf_provider.is_market_open()}")
```
