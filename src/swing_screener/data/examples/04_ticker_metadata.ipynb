# Tutorial: Ticker Metadata

Audience:
- Developers retrieving ticker metadata and applying currency logic.

Prerequisites:
- Network access for ticker-info lookups.

Learning goals:
- Fetch single and batch ticker metadata.
- Detect likely currency from ticker suffix.
- Filter a package universe by detected currency.


## Outline

1. Setup metadata and currency helpers
2. Retrieve single and multiple ticker info
3. Run suffix-based currency detection
4. Apply currency filtering on a packaged universe


```python
from swing_screener.data import load_universe_from_package
from swing_screener.data.currency import detect_currency
from swing_screener.data.ticker_info import get_multiple_ticker_info, get_ticker_info
```

## Step 1 - Fetch single ticker metadata

Start with one European and one US ticker.


```python
info_asml = get_ticker_info("ASML")
info_aapl = get_ticker_info("AAPL")

print(f"ASML: {info_asml}")
print(f"AAPL: {info_aapl}")
```

## Step 2 - Fetch metadata in batch

Use one call for multiple symbols and print selected fields.


```python
tickers = ["MSFT", "GOOGL", "NVDA", "TSLA", "META"]
results = get_multiple_ticker_info(tickers)

for ticker, data in results.items():
    print(f"{ticker}: {data['name']} | {data['sector']} | {data['currency']}")
```

## Step 3 - Detect currency from ticker suffix

This heuristic maps known exchange suffixes (for example `.AS`, `.DE`, `.PA`).


```python
test_tickers = [
    "AAPL",
    "ASML.AS",
    "SAP.DE",
    "LVMH.PA",
    "NFLX",
    "SONY.T",
]

for t in test_tickers:
    print(f"{t} -> {detect_currency(t)}")
```

## Step 4 - Filter a packaged universe by detected currency

Use `eur_all` and split symbols based on `detect_currency` output.


```python
all_eur = load_universe_from_package("eur_all")
eur_tickers = [t for t in all_eur if detect_currency(t) == "EUR"]
usd_tickers = [t for t in all_eur if detect_currency(t) == "USD"]

print(f"EUR universe: {len(eur_tickers)} EUR tickers, {len(usd_tickers)} USD tickers")
```
