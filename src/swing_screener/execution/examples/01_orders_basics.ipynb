```python
from pathlib import Path

from swing_screener.execution import Order, load_orders, save_orders
```

## Step 1 - Create pending entry orders

This mirrors the original script's starting state with two pending limit-buy entries.


```python
orders = [
    Order(
        order_id="ORD-001",
        ticker="AAPL",
        status="pending",
        order_type="BUY_LIMIT",
        quantity=100,
        limit_price=175.00,
        order_date="2024-01-15",
        notes="Breakout entry",
        order_kind="entry",
    ),
    Order(
        order_id="ORD-002",
        ticker="NVDA",
        status="pending",
        order_type="BUY_LIMIT",
        quantity=50,
        limit_price=480.00,
        order_date="2024-01-15",
        notes="Pullback entry",
        order_kind="entry",
    ),
]

for o in orders:
    print(f"{o.order_id}: {o.ticker} {o.order_type} {o.quantity} @ ${o.limit_price}")
```

## Step 2 - Save and load from JSON

Use `save_orders` and `load_orders` to confirm serialization round-trips correctly.


```python
example_path = "example_orders.json"
save_orders(example_path, orders)

loaded = load_orders(example_path)
print(f"Loaded {len(loaded)} orders")
for o in loaded:
    print(f"{o.order_id}: {o.ticker} {o.status} {o.order_type}")
```

## Step 3 - Update status and clean up

Mark the first order as filled, persist the change, and remove the temporary JSON file.


```python
loaded[0].status = "filled"
loaded[0].filled_date = "2024-01-16"
loaded[0].entry_price = 175.50
save_orders(example_path, loaded)

Path(example_path).unlink(missing_ok=True)
print("Updated first order and removed temporary file")
```
