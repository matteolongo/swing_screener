# Tutorial: Execution Guidance

Audience:
- Developers generating execution hints from screener output.

Prerequisites:
- `pandas` installed and basic DataFrame familiarity.

Learning goals:
- Build sample screener rows.
- Apply default `ExecutionConfig` behavior.
- Compare outputs under custom configuration options.


## Outline

1. Setup imports
2. Build mock screener data
3. Generate default execution guidance
4. Compare custom config variants


```python
import pandas as pd

from swing_screener.execution.guidance import ExecutionConfig, add_execution_guidance
```

## Step 1 - Create a mock screener DataFrame

These fields match the inputs expected by the guidance helper.


```python
data = {
    "ticker": ["AAPL", "NVDA", "MSFT", "GOOGL"],
    "signal": ["breakout", "pullback", "both", "breakout"],
    "last": [178.50, 485.00, 375.00, 142.00],
    "breakout_level": [177.00, 480.00, 370.00, 140.00],
    "ma20_level": [175.00, 470.00, 365.00, 138.00],
    "atr14": [3.5, 12.0, 5.0, 2.8],
}
df = pd.DataFrame(data)
df
```

## Step 2 - Run guidance with default config

Start with defaults to see suggested order types, prices, and notes.


```python
default_cfg = ExecutionConfig()
default_result = add_execution_guidance(df, default_cfg)
default_result[[
    "ticker",
    "signal",
    "suggested_order_type",
    "suggested_order_price",
    "execution_note",
]]
```

## Step 3 - Variant: disable second-chance breakout

This makes breakout handling stricter for instruments that miss the first trigger.


```python
strict_cfg = ExecutionConfig(allow_second_chance_breakout=False)
strict_result = add_execution_guidance(df, strict_cfg)
strict_result[[
    "ticker",
    "signal",
    "suggested_order_type",
    "execution_note",
]]
```

## Step 4 - Variant: tighter pullback bands

Tune pullback fractions to narrow entry bands and inspect the resulting price range.


```python
tight_cfg = ExecutionConfig(
    pullback_atr_fraction=0.15,
    pullback_band_atr_low=0.30,
)
tight_result = add_execution_guidance(df, tight_cfg)
tight_result[[
    "ticker",
    "signal",
    "suggested_order_price",
    "order_price_band_low",
    "order_price_band_high",
]]
```
